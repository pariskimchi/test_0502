먼저 실무에서 사용하는 데이터 규모를 고려해보면, AWS 클라우드 환경에서\
EMR 클러스터 상에서 ETL 파이프라인을 구축하는 코드 파일을 실행하는 방법이 알맞는 방법이라고 생각한다. 비용적인 측면을 고려해서 원천 데이터셋은 AWS S3에 적재시켜 놓고\
redshift 클러스터 내부에서 s3에 있는 데이터셋을 복사 한 후에 \
redshift 클러스터에서 전처리를 통하여 정규화된 테이블 및 분석용으로 자주 쓰일 테이블을 생성하는게 알맞는 방법이라고 생각한다.

이 모든 과정이 주간 단위로 매번 새로운 데이터가 업데이트 되고\
새로운 데이터가 업데이트 되면서 새로운 데이터에서 자동적으로 모든 과정이 진행되어야 하고,\
과정 속에서의 모니터링이 필요하기 떄문에\
Apache airflow 웹 어플리케이션을 이용하여\
redshift 클러스터와 연결시켜서 사용할것 같다

대략적으로
S3에서 Redshift로 원천 데이터셋을 복사하는 dag 파일,\
그리고 redshift 상에서의 원천 데이터셋을 이용하여 정규화된 테이블 및 여러 테이블 생성을 위한 DAG 파일이 필요할것이고,\
마지막으로 생성된 테이블들의 데이터를 확인하는 DAG 파일을 작성하는 구조를 고려해보았다
